```python
from prody import *
import pandas as pd
```

```python
df_cdk2 = pd.read_csv("./B_DATOS/TABLA_MTDATA_CDK2_402_crys.csv", index_col = 0,
                     converters={"NameLigs": lambda x: x.strip("[]").split(", ")}, 
                      # Convierte en lista la celda del nombre de los ligandos
                     )
pdb_ids = df_cdk2["PDB_ID"]
len(pdb_ids)
```

```python
from itertools import chain
pisiani_rangeResidues = [ [4,12], [17, 24], [29,34], [46,55], [66,71], [76,81],  
                         [87,93], [101, 120], [121, 135], [140, 150], [182, 194], [277, 282]]
lista = [list( range(valor[0], valor[1] + 1) ) for valor in pisiani_rangeResidues]
# Se obtiene la lista de residuos, incluida en formato de cadena de texto
pisiani_residues = list(chain.from_iterable(lista))
pisiani_residues_str = ' '.join(str(e) for e in pisiani_residues)
```

```python
# Debemos alinear las estructuras generadas con modeller a las estructuras de cadena descargadas

dir_modeller_IN = "../ARCHIVOS/CRISTALES/PROT_CDK2_MODELOS_modeller_NOPREP_TEMP/"
dir_cadenas = "../ARCHIVOS/CRISTALES/PROT_CDK2_CHAINS/"
dir_modeller_OUT = "../ARCHIVOS/CRISTALES/PROT_CDK2_MODELOS_modeller_NoPrep/"

# Leer la estructura de la cadena
# Leer la estructura modelada
# Alinear la modelada a la de la cadena usando pisiani
# 
```

```python
for pdb_id in pdb_ids:
    protein_ref = parsePDB(dir_cadenas + "/" + pdb_id + "_A.pdb")
    protein_modeller_IN = parsePDB(dir_modeller_IN + "/" + pdb_id + "_MODLL.pdb")
    protein_pocket = protein_ref.select("resid " + pisiani_residues_str )

    pdb_alg = matchAlign(protein_modeller_IN, protein_ref, overlap=70,
                                 tarsel='calpha and resnum' + ' ' + pisiani_residues_str) 
    out_protein = pdb_alg[0]
    writePDB(dir_modeller_OUT + pdb_id + '_MODLL.pdb', out_protein)
```
