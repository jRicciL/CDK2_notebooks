```python
import pdb2pqr
import pdb4amber
import os
from pathlib import Path
import glob
```

```python
Proteinas para ensamble y para docking y DM

Para ensamble
Modelos_Modeller -> pdb4amber, añadir átomos faltantes e hidrogenos, todas las histidinas quedan en HID

Modelos para docking
Modelos_Modeller -> pdb4amber sin hidrógenos, pero por sí sola corrige los enlaces 
-> pdb2pqr añade hidrógenos -> pdb4amber
Esto permite añadir 
```

```python
# Directorio actual de trabajo
cwd = os.getcwd()
out_dir_ens = "../ARCHIVOS/CRISTALES/PROT_CDK2_PREPARADAS/PREP_ALL_ENSAMBLE/" # directorio con prots histidinas protonadas según el pka
out_dir_pka = "../ARCHIVOS/CRISTALES/PROT_CDK2_PREPARADAS/PREP_PH_7/" # directorio

# Lista de archivos pdb para preparar, originalmente modelos generados con Modeller
pdb_list = glob.glob("../ARCHIVOS/CRISTALES/PROT_CDK2_MODELOS_modeller_NoPrep/*")

for pdb_file in pdb_list[0:2]:
    print(pdb_file)
    pdb_id = Path(pdb_file).stem.rsplit('_', 1)[0]
    # Para el ensamble
    ensamble_file = out_dir_ens + "/" + pdb_id + "_ENS.pdb"
    pdb4amber.run(arg_pdbin=pdb_file,
                  add_missing_atoms = True,
                  arg_pdbout = ensamble_file)
    # Para proteína a un pH 7, primero removemos los hidrógenos
    no_hydrogens_file = out_dir_pka + "/" + pdb_id + "_TEMP.pdb"
    pdb4amber.run(arg_pdbin = ensamble_file,
                  arg_nohyd = True, arg_dry = True,
                  arg_pdbout = no_hydrogens_file)
    # Ejecuta pdb2pqr para obtener los estados de protonación de los residuos ionizables
    pqr_file = out_dir_pka + "/" + pdb_id + "_PQR.pdb"
    os.system("pdb2pqr_cli --ff=amber --with-ph=7.0 --ffout=amber --ph-calc-method=propka --chain"
              + no_hydrogens_file + " " 
              + pqr_file )

    # Vuelve a ejecutar pdb4amber para generar las PROT
    prepared_file = out_dir_pka + "/" + pdb_id + "_PROT.pdb"
    pdb4amber.run(arg_pdbin= pqr_file,
                  arg_pdbout = prepared_file )

    # Elimina los archivos temporales
    #rm_files = glob.glob(cwd + "/" + out_dir_ensamble + "/*_dk_*")
    #for i in rm_files:
     #   os.remove(i)
    #os.remove(cwd + "/" + out_dir + "/" + pdb_id + "_PROT.pdb.propka")
```

```python
# Directorio actual de trabajo
cwd = os.getcwd()
out_dir = "prot_DOCKING_pka" # directorio con prots histidinas protonadas según el pka
out_dir_ensamble = "prot_ENSAMBLE_noPka" # directorio

# Lista de archivos pdb para preparar, originalmente modelos generados con Modeller
pdb_list = glob.glob("../Modelos_Full_CDK2_PDB/*")

for pdb_file in pdb_list:
    print(pdb_file)
    pdb_id = Path(pdb_file).stem.rsplit('_', 1)[0] 
    pdb4amber.run(arg_pdbin=pdb_file,
                  arg_nohyd = True, arg_dry = True,
                  arg_add_missing_atoms = True,
                  arg_pdbout = out_dir_ensamble + "/" + pdb_id + "_PROT_ens.pdb")
    
    # Elimina los archivos temporales
    rm_files = glob.glob(cwd + "/" + out_dir_ensamble + "/*_ens_*")
    for i in rm_files:
        os.remove(i)
```
