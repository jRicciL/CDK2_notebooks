```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context='notebook', style='whitegrid', font_scale=1.2)
```

```python
def _plot(var, varName, figsize=(15, 6), bins=20):
    fig, ax = plt.subplots(1, 2, figsize=figsize)
    # Histogram
    ax[0].hist(var, bins=bins)
    ax[0].set_title(f'CDK2: {varName} histogram')

    # Barplot
    var.round(2).value_counts().plot(kind='bar', ax=ax[1])
    ax[1].set_title(f'CDK2: {varName} barplot')
    plt.show()
```

## CDK2 protein

```python
_plot(df_prot.Identity/298*100, "Identity")
```

```python
df_prot.sort_values('Identity')
```

```python

_plot(df_prot.Coverage, "Coverage", figsize=(20,6))
```

```python
_plot(df_prot.Resolution.round(1), 'Resolution')
```

```python
_plot(df_prot.Inhib_mass.round(2), 'Inhibitor')
```
